```python
#importing libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#importing warnings:
import warnings
warnings.filterwarnings('ignore')

#display all the rows and columns:
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
#importing train dataset:
train=pd.read_csv(r'C:\Users\VISHAL\Web Deployment\Cloud Condition Prediction\data\exported_data\Upsample_train_data.csv',
                  index_col=0,header=0)
print(train.shape)
train.head(10)
```

```python
#importing test dataset:
test=pd.read_csv(r'C:\Users\VISHAL\Web Deployment\Cloud Condition Prediction\data\exported_data\Upsample_test_data.csv',
                  index_col=0,header=0)
print(test.shape)
test.head(10)
```

```python
#storing categorical value in 'df1_categorical' and numerical in 'df1_numerical'
train_categorical=[]
train_numerical=[]

for x in train:
    if train[x].dtype=='object':
        train_categorical.append(x)
    elif train[x].dtype=='int64' or train[x].dtype=='float64':
        train_numerical.append(x)
```

```python
train_categorical
```

```python
# For preprocessing the data 
from sklearn import preprocessing 
le=preprocessing.LabelEncoder()

for i in train_categorical:
    train[i]=le.fit_transform(train[i])
    
    le_name_mapping=dict(zip(le.classes_, le.transform(le.classes_)))
    print('Feature',i)
    print('le_name_mapping',le_name_mapping)
    print()
```

```python
#storing categorical value in 'df1_categorical' and numerical in 'df1_numerical'
test_categorical=[]
test_numerical=[]

for x in test:
    if test[x].dtype=='object':
        test_categorical.append(x)
    elif test[x].dtype=='int64' or test[x].dtype=='float64':
        test_numerical.append(x)
```

```python
test_categorical
```

```python
# For preprocessing the data 
from sklearn import preprocessing 
le=preprocessing.LabelEncoder()

for i in test_categorical:
    test[i]=le.fit_transform(test[i])
    
    le_name_mapping=dict(zip(le.classes_, le.transform(le.classes_)))
    print('Feature',i)
    print('le_name_mapping',le_name_mapping)
    print()
```
