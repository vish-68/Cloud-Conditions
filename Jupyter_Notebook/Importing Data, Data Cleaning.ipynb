```python
train_df=train_df[['Rain_OR_SNOW', 'Temperature (C)',
       'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)',
       'Wind Bearing (degrees)', 'Visibility (km)', 'Pressure (millibars)',
       'Condensation', 'Solar irradiance intensity','Cloud_Condition']]
print(train_df.shape)
train_df.head()
```

```python
#importing test dataset:
test_df=pd.read_csv(r'C:\Users\VISHAL\Web Deployment\Cloud Condition Prediction\data\test_CloudCondition.csv',
                    index_col=0,header=0)
print(test_df.shape)
test_df.head()
```

```python
train_df['Temperature (C)'].unique()
```

```python
train_df['Temperature (C)'].replace({'-':np.nan},inplace=True)
print(train_df['Temperature (C)'].unique())
print(train_df['Temperature (C)'].dtype)
```

```python
train_df["Temperature (C)"]=train_df["Temperature (C)"].astype(float)
print('Temperature (C) :',train_df['Temperature (C)'].dtype)
```

```python
train_df['Rain_OR_SNOW'].fillna(train_df['Rain_OR_SNOW'].mode()[0],inplace=True)
train_df['Temperature (C)'].fillna(train_df['Temperature (C)'].mean(),inplace=True)
train_df['Apparent Temperature (C)'].fillna(train_df['Apparent Temperature (C)'].mean(),inplace=True)
train_df['Humidity'].fillna(train_df['Humidity'].mean(),inplace=True)
train_df['Wind Speed (km/h)'].fillna(train_df['Wind Speed (km/h)'].mean(),inplace=True)
train_df['Wind Bearing (degrees)'].fillna(train_df['Wind Bearing (degrees)'].mean(),inplace=True)
train_df['Visibility (km)'].fillna(train_df['Visibility (km)'].mean(),inplace=True)
train_df['Pressure (millibars)'].fillna(train_df['Pressure (millibars)'].mean(),inplace=True)
```
